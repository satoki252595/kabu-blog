```python
##5月優待銘柄以外（クロス取引対策）で空売り比率が高めの銘柄を抽出

from method.getCreditBalance import getCreditBalance
import pandas as pd

df_1, _ = getCreditBalance()

df_1['div'] = df_1['buyBalance']/df_1['sellBalance']
df_1 = df_1[df_1['div'] < 0.5]

from method.getStockInfo import getStockInfo
from method.getStocCodes import getStockCodes

df_stockCodes = getStockCodes()
df_stockCodes.index = df_stockCodes['コード']

df = df_1.join(df_stockCodes)
df = df.dropna(how='any')

from method.getYutai import  getYuutaiCodeInformation
df_yuutai = getYuutaiCodeInformation('may')
df['優待有無'] = False

for code in df.index:
    if str(code) in df_yuutai.keys():
        df.at[code,'優待有無']=True

df = df[df['優待有無']==False]
drop_col = ['日付','コード','市場・商品区分','33業種コード','17業種コード','17業種区分','規模コード','規模区分','優待有無']
df = df.drop(drop_col, axis=1)
df = df.sort_values(by="div")

from method.getStockInfo import getStockInfo

df['avg20_price'] = 0
df['avg20_volume'] = 0

for code in df.index:
    df_stockInfo = getStockInfo.getStockPriceInfo(str(code),period = '20d')
    df.at[code,'avg20_price'] = df_stockInfo['Close'].mean()
    df.at[code,'avg20_volume'] = df_stockInfo['Volume'].mean()
```
